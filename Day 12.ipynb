Customizing Plot Legends	
    Choosing Elements for the Legend
	Legend for Size of Points
	Multiple Legends
	Customizing Colorbars

```python
#legend
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-', label='Sine')
ax.plot(x, np.cos(x), '--', label='Cosine')
ax.legend(loc='upper right',frameon=False,ncol=1,borderpad=1, fancybox=True, shadow=True)
```

```python
#Multiple legend

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
lines=[]
styles=['-', '--', '-.', ':']
for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='black')
leg = ax.legend()
ax.legend(lines[:2], ['line A', 'line B'],loc='upper right', frameon=True)

from matplotlib.legend import Legend
l = Legend(ax, lines[2:], ['line C', 'line D'],loc='lower right', frameon=True)
ax.add_artist(l) #--> add_artist
```

Multiple Subplots	
    plt.axes: Subplots by Hand
	plt.subplot: Simple Grids of Subplots
	plt.subplots: The Whole Grid 
	plt.GridSpec: More Complicated Arrangements

```python
#  plt.axes: Subplots by Hand (left , bottom , width , height)

import matplotlib.pyplot as plt
import numpy as np
f=plt.figure()
a1=f.add_axes([0.1,0.5,0.8,0.4])
a2=f.add_axes([0.1,0.1,0.8,0.4])
x=np.linspace(0,10,100)
a1.plot(np.sin(x))
a2.plot(np.cos(x))
```

```python
#plt.subplot: Simple Grids of Subplots

#plt.subplot(row, col , panel num) 

for i in range(1,8):   # 7 plots --> 3, 3 
    plt.subplot(3,3,i)
    plt.text(0.5,0.5,str((3,3,i)),fontsize=10)
```

```python
#plt.subplots: The Whole Grid  (row, col)

f,a= plt.subplots(4,5,sharex='col', sharey='row')
for i in range(4):
    for j in range(5):
        a[i,j].text(0.5,0.5, str((i,j)),fontsize=6)
```

```python
#plt.GridSpec: More Complicated Arrangements + plt.subplot

grid= plt.GridSpec(2,3,wspace=0.4, hspace=0.4)
plt.subplot(grid[0,0])
plt.subplot(grid[0,1])
plt.subplot(grid[0,2])
plt.subplot(grid[1,0])
plt.subplot(grid[1,1])
plt.subplot(grid[1,2])
```

Text and Annotation	
    1.Transforms and Text Position ->axes.transData, axes.Transaxes, f.transfigure
	2.Arrows and Annotation -> plt.arrow(), plt.annotate()

```python
f,a = plt.subplots()
a.axis([0,10,0,10])
a.text(2,6,'(2,6)',transform=a.transData) #text(row,col, string , tranform, fontsize)
a.text(0.5,0.7,"(0.5,0.7)",transform=a.transAxes)
a.text(0.2,0.2,"(0.2,0.2)",transform=f.transFigure)
```

```python
#2.Arrows and Annotation -> plt.arrow(), plt.annotate()
#https://matplotlib.org/stable/_images/sphx_glr_fancyarrow_demo_001.png
import matplotlib
f,a= plt.subplots()
x=np.linspace(0,20,100)
a.plot(x,np.cos(x))
a.axis('equal')
a.annotate('Local Max', xy=(5.7,1.5), xytext=(10,4), arrowprops=dict())
a.annotate('Local Min', xy=(9.5,-1.5), xytext=(2,-6),arrowprops=dict(arrowstyle='->'))
```
