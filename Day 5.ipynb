```python
 #np.sort ->  Quick sort ->O[N LOg N]
    
import numpy as np
a=np.array([8,5,3,7,10])
print(np.sort(a)) 
print(np.argsort(a)) # sorting index  2,1,3,0,4
print(np.partition(a,1)) #partial sorting
print(np.argpartition(a,1)) # return index of partial sorting
```

```python
#K-Nearest Neighbors: -> Nearest points 

import matplotlib.pyplot as plt
import seaborn; seaborn.set()
rand = np.random.RandomState(42)
X = rand.rand(10, 2)
plt.scatter(X[:, 0], X[:, 1]);
differences = X[:, np.newaxis, :] - X[np.newaxis, :, :]
print(differences.shape)
sq_differences = differences ** 2
print(sq_differences.shape)
dist_sq = sq_differences.sum(-1)
print(dist_sq.shape)
print(dist_sq.diagonal())
print(dist_sq)
nearest = np.argsort(dist_sq, axis=1)
print(nearest)
K=2 
nearest_partition = np.argpartition(dist_sq, K + 1, axis=1)
print(nearest_partition)
```

```python
K=2
for i in range(X.shape[0]):
    for j in nearest_partition[i, :K+1]:
          plt.plot(*zip(X[j], X[i]), color='black')
```

	Advanced Compound Types
	Structured Arrays

```python
#Creating Structured Arrays
import numpy as np
name=np.array(['a','b','c'])
age=np.array([1,2,3])
weight=np.array([20.0,30.5,18.5])
#empty numpy array
data= np.zeros(3, dtype={'names':('name','age','weight'),
                         'formats':('U10','i4','f8')})

print(data.dtype)
data['name']=name
data['age']=age
data['weight']=weight
print(data)
#import unicodedata
#U -> String data type  U10 
#i->integers i4 
#f-> float f8 -> 8 bits of  
```

```python
#Advanced Compound Types

#Matrix -> Numpy 

Ac= np.dtype([('id', 'i8'), ('mat', 'f8', (3, 3))])
X=np.zeros(1,dtype=Ac)
print(X)
print(X['id'])
print(X['mat'])
```
