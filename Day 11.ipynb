```python
# Adjusting the Plot: Line Colors and Styles 
 #Line Colors
#1.By name (green, red,blue )
#2.by hex code eg: #000000 ->black #ffffff ->white  # refer https://www.color-hex.com/color/ffffff
#3.Gray scale code (0.75) 0 to 1
#4.by short rgb (rgbcmyk) (g, r, b, y ,v)

#eg: plt.plot(x, np.sin(x),color='green')
    
    
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=np.array([2,3,4,6,8,9])
y=np.array([4,5,6,7,8,9])
#plt.plot(x,y, color='green')



# Line styles : dashed, solid, dotted, dashdot
# '-' =>solid
# '--' =>dashed
# '-.' =>dashdot
# ':' =>Dotted
# '' =>None
# 'o'
plt.plot(x,y, color='green', linestyle='-.')
```

```python
#Adjusting the Plot: Axes Limits

#xlim(0,6)
#ylim(1,10)

plt.xlim(0,10)
plt.ylim(0,10)
plt.plot(x, np.sin(x), color='#000000', linestyle='solid')

#Bounding concept

#plt.axis('tight') 
#plt.axis('equal')
```

```python
#Labeling Plots:
#1.Title
#2.Label , Legend
#3.X, Y axis

plt.plot(x, np.sin(x), color='#000000', linestyle='solid' , label='sin(x)')
plt.plot(x, np.cos(x), color='blue', linestyle='solid' , label='cos(x)')
plt.title('Sin Black solid wave', fontsize=10,color='Green')
plt.xlabel('x')
plt.ylabel('Sin & Cos Wave')
plt.legend()
```

Simple Scatter Plots	
    Scatter Plots with plt.plot -->  Plot is more efficient than scatter
	Scatter Plots with plt.scatter --> Scatter will take much time to load the all features
	plot Versus scatter

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=np.linspace(0,10,40)
y=np.sin(x)
plt.plot(x,y,'o')   
```

```python
rng=np.random.RandomState(0)
for i in ['o','.','+','-','*','v','d','s','k']:
    plt.plot(rng.rand(5),rng.rand(5),i,label="marker='{0}'".format(i))
plt.legend()
```

```python
#plt.scatter

x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x,y)
```

```python
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
colors=rng.rand(100)
sizes=1000*rng.rand(100)
plt.scatter(x,y,c=colors,s=sizes)
plt.colorbar() #--> it denotes the color in plotting
```
