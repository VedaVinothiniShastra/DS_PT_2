```python
#Exploring Seaborn Plots -> hist , kde, density


data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
d=pd.DataFrame(data, columns=['x', 'y'])
for col in 'xy':
    plt.hist(d[col], alpha=0.5)
```

```python
#KDE -> sb.kdeplot

import seaborn as sb
for col in 'xy':
    sb.kdeplot(d[col], shade=True)
```

```python
#adding two plots -> hist+kde

sb.distplot(d['x'])
sb.distplot(d['y'])
```

```python
#joinplot -> sb.joinplot

#--> kind --> kde, scatter , hex
with sb.axes_style('white'):
    sb.jointplot( d, kind='kde'); 

with sb.axes_style('white'):
    sb.jointplot( d, kind='scatter'); 

with sb.axes_style('white'):
    sb.jointplot( d,x='x', y='y', kind='hex'); 
```

```python
#Classification
#Scikit-Learn , pandas ,numpy , matplotlib , seaborn 
from sklearn.datasets import make_blobs #--> To create a data set
from sklearn.svm import SVC #Support vector classifier
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#Step 1 -> Creating a blob for classifier prblm for training data
X, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60) 
#Step 2 -> Applying Algo
clf = SVC(kernel='linear')
#Step 3 -> Fitting
clf.fit(X,y)


#Step 4 -> create a testing data

X2,_ = make_blobs(n_samples=80, centers=2,
                  random_state=0, cluster_std=0.80) 

X2=X2[50:]

#Step 5 ---> Predicting

y2=clf.predict(X2)


#Step 6 --> Visualization

#for Training Data 

#fig, ax = plt.subplots(figsize=(8, 6))
#point_style = dict(cmap='Paired', s=50)
#ax.scatter(X[:, 0], X[:, 1], c=y, **point_style)


fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

#Visualization of lines in X2 data 
xx = np.linspace(-1, 4, 10)
yy = np.linspace(-2, 7, 10)
xy1, xy2 = np.meshgrid(xx, yy)
Z = np.array([clf.decision_function([t])
              for t in zip(xy1.flat, xy2.flat)]).reshape(xy1.shape)
line_style = dict(levels = [-1.0, 0.0, 1.0],
                  linestyles = ['dashed', 'solid', 'dashed'],
                  colors = 'gray', linewidths=1)



point_style = dict(cmap='Paired', s=50)

#For Test data X2
ax[0].scatter(X2[:, 0], X2[:, 1], c='gray', **point_style)
ax[0].axis([-1, 4, -2, 7])

#For Test data X2 with Line
ax[1].scatter(X2[:, 0], X2[:, 1], c=y2, **point_style)
ax[1].contour(xy1, xy2, Z, **line_style)
ax[1].axis([-1, 4, -2, 7])
```
